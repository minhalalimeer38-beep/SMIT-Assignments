```python
# Pandas library import kar rahe hain data handle karne ke liye
import pandas as pd
```

```python
# CSV file ko read karke dataframe me store kar rahe hain
ds = pd.read_csv("student.csv")
```

```python
ds
```

```python
# Dataset ki basic information dekhne ke liye (columns, data types, null values)
ds.info()
```

```python
# Duplicate rows check karne ke liye
ds.duplicated().sum()
```

```python
# Duplicates Data ko drop karne ke liye
ds.drop_duplicates(inplace = True)
```

```python
ds
```

```python
# Extracurricular Activities column me yes aur no ko numbers me convert kar rahe hain
# YES = 1, NO = 0
ds["Extracurricular Activities"] = ds["Extracurricular Activities"].replace(["Yes" , "No"] , [1 , 0])
```

```python
ds
```

```python
# columns ka name change kar rahy hain apni asani ke liye
ds.rename(columns={"Hours Studied" : "hour_studied" , 
                   "Previous Scores" : "previous_scores" , 
                   "Extracurricular Activities" : "extra_activities" , 
                   "Sleep Hours" : "sleep_hours" , 
                   "Sample Question Papers Practiced" : "past_paper_practiced" ,
                   "Performance Index" : "performance"} , inplace = True)
```

```python
ds
```

```python
# Independent variables (features) select kar rahe hain (Performance ke ilawa sab)
x = ds.drop("performance" , axis = 1)

# Dependent variables (Target) select kar rahe hain (Only Performance)
y = ds["performance"]
```

```python
# Dataset ko training aur testing me split karne ke liye function import kar rahe hain
from sklearn.model_selection import train_test_split

# 50% Training aur 50% Testing data bana rahe hain
X_train , X_test , y_train , y_test = train_test_split (x , y , test_size = 0.50 , random_state = 62)
```

```python
# Linear Regression model import kar rahe hain
from sklearn.linear_model import LinearRegression
```

```python
# Model create kar rahe hain
model = LinearRegression()

# Training data par model train (fit) kar rahe hain
model.fit(X_train , y_train)
```

```python
# Model ka Coefficient dekhne ke liye
model.coef_
```

```python
# Model ka Interceopt dekhny ke liye
model.intercept_
```

```python
# Training data par predictions nikal rahe hain
y_pred_train = model.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error , r2_score
from math import sqrt
```

```python
# Training data ka Mean Squared Error calculate kar rahe hain
mean_squared_error(y_train , y_pred_train)
```

```python
# Training data ka Root Mean Squared Error calculate kar rahe hain
sqrt(mean_squared_error(y_train , y_pred_train))
```

```python
# Training data ka R2_Score calculate kar rahe hain
r2_score(y_train , y_pred_train)
```

```python
# Testing data par predictions nikal rahe hain
y_pred_test = model.predict(X_test)
```

```python
# Testing data ka Mean Squared Error calculate kar rahe hain
mean_squared_error(y_test , y_pred_test)
```

```python
# Testing data ka root Mean Squared Error calculate kar rahe hain
sqrt(mean_squared_error(y_test , y_pred_test))
```

```python
# Testing data ka R2_Score calculate kar rahe hain
r2_score(y_test , y_pred_test)
```

```python
# User se Input le rahy hein
Hours_Studied         = float(input("Enter Hours Studied : "))
Previous_Scores      = float(input("Enter Previous Scores : "))
Extra_Activities     = int(input("Enter Extra Activities 1 or 0 : "))
Sleep_Hours          = float(input("Enter Sleep Hour : "))
Past_Paper_Practiced = int(input("Enter Past Paper Practiced 1 or 0 : "))

# Function bana rahy hein Charges Pridect karne ke liye
def Performance_Predictor(hours_studied , previous_scores , extra_activities , sleep_hours , past_paper_practiced):
    Performance = model.predict([[hours_studied , previous_scores , extra_activities , sleep_hours , past_paper_practiced]])
    return Performance
result = Performance_Predictor(Hours_Studied , Previous_Scores , Extra_Activities , Sleep_Hours , Past_Paper_Practiced)
print(result)
```
