```python
# Pandas library import kar rahe hain data handle karne ke liye
import pandas as pd
```

```python
# CSV file ko read karke dataframe me store kar rahe hain
dh = pd.read_csv("house.csv")
```

```python
dh
```

```python
# Dataset ki basic information dekhne ke liye (columns, data types, null values)
dh.info()
```

```python
# Duplicate rows check karne ke liye
dh.duplicated().sum()
```

```python
# columns ka name change kar rahy hain apni asani ke liye
dh = dh.rename(columns={
    "X1 transaction date": "transaction_date",
    "X2 house age": "house_age",
    "X3 distance to the nearest MRT station": "mrt_distance",
    "X4 number of convenience stores": "convenience_stores",
    "X5 latitude": "latitude",
    "X6 longitude": "longitude",
    "Y house price of unit area": "price"
})
```

```python
dh
```

```python
# No ko drop kar rahy hein q k iski zaroorat nahi
dh = dh.drop("No" ,axis = 1)
```

```python
dh
```

```python
# Independent variables (features) select kar rahe hain (price ke ilawa sab)
x = dh.drop("price" , axis = 1)

# Dependent variables (Target) select kar rahe hain (Only price)
y = dh["price"]
```

```python
# Dataset ko training aur testing me split karne ke liye function import kar rahe hain
from sklearn.model_selection import train_test_split

# 70% Training aur 30% Testing data bana rahe hain
X_train , X_test , y_train , y_test = train_test_split (x , y , test_size = 0.3 , random_state = 42)
```

```python
# Linear Regression model import kar rahe hain
from sklearn.linear_model import LinearRegression
```

```python
# Model create kar rahe hain
model = LinearRegression()

# Training data par model train (fit) kar rahe hain
model.fit(X_train , y_train)
```

```python
# Model ka Coefficient dekhne ke liye
model.coef_
```

```python
# Model ka Interceopt dekhny ke liye
model.intercept_
```

```python
# Training data par predictions nikal rahe hain
y_pred_train = model.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error , r2_score
from math import sqrt
```

```python
# Training data ka Mean Squared Error calculate kar rahe hain
mean_squared_error(y_train , y_pred_train)
```

```python
# Training data ka Root Mean Squared Error calculate kar rahe hain
sqrt(mean_squared_error(y_train , y_pred_train))
```

```python
# Training data ka R2_Score calculate kar rahe hain
r2_score(y_train , y_pred_train)
```

```python
# Testing data par predictions nikal rahe hain
y_pred_test = model.predict(X_test)
```

```python
# Testing data ka Mean Squared Error calculate kar rahe hain
mean_squared_error(y_test , y_pred_test)
```

```python
# Testing data ka root Mean Squared Error calculate kar rahe hain
sqrt(mean_squared_error(y_test , y_pred_test))
```

```python
# Testing data ka R2_Score calculate kar rahe hain
r2_score(y_test , y_pred_test)
```

```python
# Function bana rahy hein Charges Pridect karne ke liye
def House_Price_Predictor(transaction_date , house_age , mrt_distance , convenience_stores , latitude , longitude):
    price = model.predict([[transaction_date , house_age , mrt_distance , convenience_stores , latitude , longitude]])
    return (price)

# User se Input le rahy hein
Transaction_Date   = float(input("Enter Transaction Date : "))
House_Age          = float(input("Enter House Age : "))
MRT_Distance       = float(input("Enter MRT Distance : "))
Convenience_Stores = float(input("Enter Convenience Stores : "))
Latitude           = float(input("Enter Latitude : "))
Longitude          = float(input("Enter Longitude : "))

result = House_Price_Predictor(Transaction_Date , House_Age , MRT_Distance , Convenience_Stores , Latitude , Longitude)
print(result)
```
