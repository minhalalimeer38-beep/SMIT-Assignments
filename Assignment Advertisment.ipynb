```python
# Pandas library import kar rahe hain data handle karne ke liye
import pandas as pd
```

```python
# CSV file ko read karke dataframe me store kar rahe hain
da = pd.read_csv("advertisment.csv")
```

```python
da
```

```python
# Dataset ki basic information dekhne ke liye (columns, data types, null values)
da.info()
```

```python
# Duplicate rows check karne ke liye
da.duplicated().sum()
```

```python
# Independent variables (features) select kar rahe hain (Sales ke ilawa sab)
x = da.drop("Sales" , axis = 1)

# Dependent variable (target) select kar rahe hain (Sales)
y = da.Sales
```

```python
# Dataset ko training aur testing me split karne ke liye function import kar rahe hain
from sklearn.model_selection import train_test_split

# 75% Training aur 25% Testing data bana rahe hain
X_train , X_test , y_train , y_test = train_test_split(x , y , test_size = 0.25 , random_state = 42)
```

```python
# Linear Regression model import kar rahe hain
from sklearn.linear_model import LinearRegression

# Model create kar rahe hain
model = LinearRegression()

# Training data par model train (fit) kar rahe hain
model.fit(X_train , y_train)
```

```python
# Model ke coefficients dekhne ke liye
model.coef_
```

```python
# Model ka intercept dekhne ke liye
model.intercept_
```

```python
# Training data par predictions nikal rahe hain
y_pred_train = model.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error , r2_score
from math import sqrt
```

```python
# Training data ka Mean Squared Error calculate kar rahe hain
mean_squared_error(y_train , y_pred_train)
```

```python
# Training data ka root Mean Squared Error calculate kar rahe hain
sqrt(mean_squared_error(y_train , y_pred_train))
```

```python
# Training data ka R2 score calculate kar rahe hain (model accuracy measure)
r2_score(y_train , y_pred_train)
```

```python
# Testing data par predictions nikal rahe hain
y_pred_test = model.predict(X_test)
```

```python
# Testing data ka Mean Squared Error calculate kar rahe hain
mean_squared_error(y_test , y_pred_test)
```

```python
# Testing data ka root Mean Squared Error calculate kar rahe hain
sqrt(mean_squared_error(y_test , y_pred_test))
```

```python
# Testing data ka R2 score calculate kar rahe hain (model accuracy measure)
r2_score(y_test , y_pred_test)
```

```python
# Function bana rahe hain jo TV, Radio aur Newspaper input lekar Sales predict kare
def Sales_Predictor(TV , Radio , Newspaper):
    sales = model.predict([[TV , Radio , Newspaper]])
    return sales

# User se inputs le rahe hain
TV        = float(input("Enter TV invest    : "))
Radio     = float(input("Enter TV Radio     : "))
Newspaper = float(input("Enter TV Newspaper : "))

# Function call karke result store kar rahe hain
result = Sales_Predictor(TV , Radio , Newspaper)
print (result)
```
