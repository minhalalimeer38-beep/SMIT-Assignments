```python
# Pandas library import kar rahe hain data handle karne ke liye
import pandas as pd
```

```python
# CSV file ko read karke dataframe me store kar rahe hain
dt = pd.read_csv("insurance.csv")
```

```python
dt
```

```python
# Dataset ki basic information dekhne ke liye (columns, data types, null values)
dt.info()
```

```python
# Duplicate rows check karne ke liye
dt.duplicated().sum()
```

```python
# Duplicates Data ko drop karne ke liye
dt = dt.drop_duplicates()
```

```python
dt
```

```python
# Sex column me male aur female ko numbers me convert kar rahe hain
# male = 1, female = 0
dt["sex"] = dt["sex"].replace(["male" , "female"] , [1 , 0])

# Smoker column me yes aur no ko numbers me convert kar rahe hain
# yes = 1, no = 0
dt.smoker = dt["smoker"].replace(["yes" , "no"] , [1 , 0])
```

```python
dt
```

```python
# Data ki Information nikalne ke liye 
dt.info()
```

```python
# Categorical variables ko numerical (0 aur 1) me convert karne ke liye get_dummies use kar rahe hain
dt = pd.get_dummies(dt  ,dtype = int)
```

```python
dt
```

```python
dt.info()
```

```python
# Independent variables (features) select kar rahe hain (Charges ke ilawa sab)
x = dt.drop("charges" , axis = 1)

# Dependent variables (Target) select kar rahe hain (Only Charges)
y = dt.charges
```

```python
# Dataset ko training aur testing me split karne ke liye function import kar rahe hain
from sklearn.model_selection import train_test_split

# 70% Training aur 30% Testing data bana rahe hain
X_train , X_test , y_train , y_test = train_test_split(x , y , test_size = 0.30 , random_state = 53)
```

```python
# Linear Regression model import kar rahe hain
from sklearn.linear_model import LinearRegression

# Model create kar rahe hain
model = LinearRegression()

# Training data par model train (fit) kar rahe hain
model.fit(X_train , y_train)
```

```python
# Model ka Coefficient dekhne ke liye
model.coef_
```

```python
# Model ka Interceopt dekhny ke liye
model.intercept_
```

```python
from sklearn.metrics import mean_squared_error , r2_score
from math import sqrt
```

```python
# Training data par predictions nikal rahe hain
y_pred_train = model.predict(X_train)
```

```python
# Training data ka Mean Squared Error calculate kar rahe hain
mean_squared_error(y_train , y_pred_train)
```

```python
# Training data ka root Mean Squared Error calculate kar rahe hain
sqrt(mean_squared_error(y_train , y_pred_train))
```

```python
# Training data ka R2_Score calculate kar rahe hain
r2_score(y_train , y_pred_train)
```

```python
# Testing data par predictions nikal rahe hain
y_pred_test = model.predict(X_test)
```

```python
# Testing data ka Mean Squared Error calculate kar rahe hain
mean_squared_error(y_test , y_pred_test)
```

```python
# Testing data ka root Mean Squared Error calculate kar rahe hain
sqrt(mean_squared_error(y_test , y_pred_test))
```

```python
# Testing data ka R2_Score calculate kar rahe hain
r2_score(y_test , y_pred_test)
```

```python
# Function bana rahy hein Charges Pridect karne ke liye
def Charges_predictor(age , sex , bmi , child , smoker , northwest , southeast , northeast , southwest):
    charges = model.predict([[age , sex , bmi , child , smoker , northwest , southeast , northeast , southwest]])
    return charges

# User se Input le rahy hein
Sex       = int(input("Enter Sex 1 or 0 :"))
Age       = int(input("Enter Age :"))
BMI       = float(input("Enter BMI :"))
Child     = int(input("Enter Children :"))
Smoker    = int(input("Enter 1 or 0 :"))
Northwest = int(input("Enter 1 or 0 :"))
Southeast = int(input("Enter 1 or 0 :"))
Northeast = int(input("Enter 1 or 0 :"))
Southwest = int(input("Enter 1 or 0 :"))

# Function call kar ke result store kar rahe hein
result = Charges_predictor(Age, Sex, BMI, Child, Smoker, Northwest, Southeast, Northeast, Southwest)
print(result)
```
